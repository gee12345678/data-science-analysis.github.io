```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Sample - Superstore.csv', encoding='ISO-8859-1')


df.head()
```

```python
# Convert date columns to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Remove duplicates (if any)
df = df.drop_duplicates()

# Preview cleaned data
df.head()
```

```python
# Check for missing values
missing = df.isnull().sum()
print("Missing values per column:\n", missing)

# If missing values existed, you could fill or drop them:
# df = df.dropna()  # drop missing rows
# df['Column'] = df['Column'].fillna(value)  # fill specific column

# Since this dataset has no missing values, no changes needed
```

```python
# Describe numeric columns
df.describe()
```

```python


# Total sales and profit by category
plt.figure(figsize=(10,5))
sns.barplot(data=df, x='Category', y='Sales', estimator=sum)
plt.title('Total Sales by Category')
plt.show()

plt.figure(figsize=(10,5))
sns.barplot(data=df, x='Category', y='Profit', estimator=sum)
plt.title('Total Profit by Category')
plt.show()
```

```python
# Boxplot for Sales
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='Category', y='Sales')
plt.title('Sales Outliers by Category')
plt.show()

# Boxplot for Profit
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='Category', y='Profit')
plt.title('Profit Outliers by Category')
plt.show()
```

```python
from scipy import stats
import numpy as np

# Filter based on Z-score
z_scores = np.abs(stats.zscore(df[['Sales', 'Profit']]))
df_clean = df[(z_scores < 3).all(axis=1)]

print(f"Original rows: {len(df)}, Cleaned rows: {len(df_clean)}")
```

```python
# Total profit by region and category
grouped = df.groupby(['Region', 'Category'])[['Sales', 'Profit']].sum().reset_index()
print(grouped)
```

```python
# Aggregate by order date
time_data = df_clean.groupby('Order Date')[['Sales', 'Profit']].sum().reset_index()

# Line plot for Sales and Profit over time
plt.figure(figsize=(14, 6))
plt.plot(time_data['Order Date'], time_data['Sales'], label='Sales')
plt.plot(time_data['Order Date'], time_data['Profit'], label='Profit')
plt.title('Sales and Profit Over Time')
plt.xlabel('Date')
plt.ylabel('Amount')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Create month-year column
df_clean['Month'] = df_clean['Order Date'].dt.to_period('M')

# Monthly summary
monthly_data = df_clean.groupby('Month')[['Sales', 'Profit']].sum().reset_index()
monthly_data['Month'] = monthly_data['Month'].astype(str)

# Plot
plt.figure(figsize=(14, 6))
sns.lineplot(data=monthly_data, x='Month', y='Sales', label='Sales')
sns.lineplot(data=monthly_data, x='Month', y='Profit', label='Profit')
plt.xticks(rotation=45)
plt.title('Monthly Sales and Profit Trend')
plt.tight_layout()
plt.show()
```

```python
# Top sub-categories by total sales
top_sales = df_clean.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).head(10)
top_profit = df_clean.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False).head(10)

# Plot
plt.figure(figsize=(12,5))
top_sales.plot(kind='bar', title='Top 10 Sub-Categories by Sales', color='skyblue')
plt.ylabel('Total Sales')
plt.show()

plt.figure(figsize=(12,5))
top_profit.plot(kind='bar', title='Top 10 Sub-Categories by Profit', color='lightgreen')
plt.ylabel('Total Profit')
plt.show()
```

```python
# Correlation matrix
correlation = df_clean[['Sales', 'Quantity', 'Discount', 'Profit']].corr()

# Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
# Create a pivot table
pivot_table = pd.pivot_table(
    df_clean,
    values=['Sales', 'Profit'],
    index='Region',
    columns='Category',
    aggfunc='sum',
    margins=True,
    margins_name='Total'
)

# Display the pivot table
pivot_table
```
